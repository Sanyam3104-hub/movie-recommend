```python
import pandas as pd 
import numpy as np 
import ast
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credit=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credit.head(1)
```

# merge the dataset

```python
movies=movies.merge(credit,on="title")
```

```python
movies.head(3)
```

 # selecting the column for our model 

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
    
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
def change(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(change)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(change)
```

```python
movies.head(1)
```

```python
def convert(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i["name"])
            counter=counter+1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
def fetch(obj):
    lists=[]
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
            lists.append(i['name'])
            break
            
    return lists
```

```python
movies['crew']=movies['crew'].apply(fetch)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_data=movies[['title','movie_id','tags']]
```

```python
new_data.head()
```

```python
new_data['tags']=new_data['tags'].apply(lambda x:" ".join(x))
```

```python
new_data
```

```python
new_data['tags']=new_data['tags'].apply(lambda x:x.lower())
```

```python
new_data
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_data['tags']=new_data['tags'].apply(stem)
```

```python
new_data['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new_data['tags']).toarray()
```

```python
print(vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
def recommend(movie):
    movie_index = new_data[new_data['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_data.iloc[i[0]].title)
     
```

```python
recommend('John Carter')
```
